```python
import pandas as pd

from bokeh.plotting import figure
from bokeh.models import ColumnDataSource
from bokeh.io import show, output_notebook
from bokeh.palettes import Category10
from bokeh.transform import factor_cmap

# for laying out the charts in either horizontal or vertical order
from bokeh.layouts import Row, Column

# import boxselect
from bokeh.models import BoxSelectTool

output_notebook()
```

```python
dfAuto = pd.read_excel('data/Auto_Insu.xlsx')
```

```python
Category10[7]
```

# Create a scatter plot of HP Vs City_MPG

```python
cds = ColumnDataSource(dfAuto)
cds.column_names
```

```python
scatter = figure(height=400,width=400,
                title="HP VS City MPG",
                x_axis_label="Horsepower",
                y_axis_label="City MPG")


scatter2 = figure(height=400,width=400,
                title="HP VS Highway MPG",
                x_axis_label="Horsepower",
                y_axis_label="Highway MPG",
                 # bneed to explicitly specify sharing of ranges to enable linked panning
                  x_range=scatter.x_range,
                 y_range=scatter.y_range)
```

```python
myColor = factor_cmap('Engine_Type',
                      palette=Category10[8],
                      factors=sorted(dfAuto['Engine_Type'].unique()) )
```

```python
# Linked brushing: make sure both charts are using the same Columndatasource

scatter.circle(x='HP',
              y='City_MPG',
              source=cds,
              color=myColor)

scatter2.x(x='HP',
              y='Highway_MPG',
              source=cds,
              color=myColor)

scatter.x_range.start=0
scatter.y_range.start=0
scatter2.x_range.start=0
scatter2.y_range.start=0

# Another approach to add tools to the figure after creating it
scatter.add_tools(BoxSelectTool())
scatter2.add_tools(BoxSelectTool())

# Laying out charts next to each other
show(Row(scatter,scatter2))
```
