```python
import pandas as pd
from bokeh.plotting import figure
from bokeh.models import ColumnDataSource
from bokeh.io import show, output_notebook
from bokeh.transform import dodge
from bokeh.core.properties import value

output_notebook()
```

```python
df = pd.read_excel('StocksClosingPrices.xlsx')
df.head()
```

```python
cds_df = ColumnDataSource(df)
cds_df.column_names
```

```python
chart = figure(height = 400, width = 600, 
               title = "Stock Price Trends", 
               x_axis_type = "datetime",
               tools = "pan,wheel_zoom,box_zoom,reset,lasso_select")
```

```python
chart.line(x = "Date", y = "Amazon", color = "darkorange", 
           legend = value("Amazon"), source = cds_df)
chart.line(x = "Date", y = "Google", color = "green",
           legend = value("Google"), source = cds_df)
chart.line(x = "Date", y = "IBM", color = "indigo",
           legend = value("IBM"), source = cds_df)
chart.legend.click_policy="hide"
show(chart)
```
