```python
data = data.rename(columns = {'% Change in Mortality Rate, 1980-2014':'mean',
                              '% Change in Mortality Rate, 1980-2014 lower':'lower',
                              '% Change in Mortality Rate, 1980-2014 upper':'upper',
                              'Location-County':'county','Location-State':'state'})
data.columns
```

```python
cod_state = data.groupby(['COD','state'])['Value'].mean()
cod_state_df = pd.DataFrame(cod_state)
cod_state_df = cod_state_df.reset_index()
cod_state_df.head()
```

```python
cod_value = cod_state_df.copy()
cod_value = cod_value.pivot(index='state',columns='COD',values='Value')
cod_value.head()
```

```python
cod_value = cod_value.reset_index()
#cod_value.to_csv(r'cod_as_variables.csv')
```

```python
cod_county = data.groupby(['COD','state','county','Year'])['Value'].mean()
cod_county_df = pd.DataFrame(cod_county)
cod_county_df = cod_county_df.reset_index()
cod_county_df.info()
```

```python
cod_county_t = cod_county_df[['state','county','COD','Year','Value']].copy()
cod_county_t = cod_county_t.pivot_table(index=['state','county','Year'],columns='COD',values='Value')
cod_county_t.info()
```

```python
cod_county_t = cod_county_t.reset_index()
cod_county_t.info()
#cod_county_t.to_csv(r'year_state_county_cod_as_variables.csv')
```
